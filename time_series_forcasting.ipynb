```python
# Outliers by z-score
train['z_score'] = (train['sales'] - train['sales'].mean()) / train['sales'].std()

outliers = train[(train['z_score'].abs() > 3)]
outliers
```

```python
# Outliers by IQR
Q1 = train['sales'].quantile(0.25)
Q3 = train['sales'].quantile(0.75)
IQR = Q3 - Q1
outliers = train[(train['sales'] < (Q1 - 1.5 * IQR)) | (train['sales'] > (Q3 + 1.5 * IQR))]
outliers
```
